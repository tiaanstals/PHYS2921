```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# set your number of slits
n = 2
```

```python
x = np.linspace(-0.04, 0.04, 1000)
# wavelength
wl = 488e-9
#slit seperation
a = 6.21e-5
# width of slits
b = 30e-6
def n_slit(n, x):
    alpha = (np.pi*a*np.sin(x))/wl
    beta = (np.pi*b*np.sin(x))/wl
    return (np.sin(beta)/beta)**2 * (np.sin(n*alpha)/(n*np.sin(alpha)))**2

def envelope(n, x):
    alpha = (np.pi*a*np.sin(x))/wl
    beta = (np.pi*b*np.sin(x))/wl
    return (np.sin(beta)/beta)**2

def other(n, x):
    alpha = (np.pi*a*np.sin(x))/wl
    beta = (np.pi*b*np.sin(x))/wl
    return (np.sin(n*alpha)/(n*np.sin(alpha)))**2

y = n_slit(n,x)
y_env = envelope(n,x)
y_other = other(n,x)
fig=plt.figure(figsize=(20,10))
plt.plot(x,y, label=r'$(\frac{sin(\beta)}{\beta})^2 (\frac{sin(N \alpha)}{N sin(\alpha)})^2$',c='b')
plt.plot(x,y_env, 'g--',label=r'$(\frac{sin(\beta)}{\beta})^2$')
plt.plot(x,y_other, 'r--',label=r'$(\frac{sin(N \alpha)}{N sin(\alpha)})^2$')
plt.legend(prop={'size': 18})
```
